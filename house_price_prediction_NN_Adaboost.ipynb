```python
#Importing the required packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
#Fetching the dataset from sklearn datasets

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
#Checking the feature names of the dataset

print(housing.feature_names[0:])
```

```python
#Checking the dimension of the data

print(housing.data.shape)
```

```python
#Checking the dimension of the target

print(housing.target.shape)
```

```python
#Converting the dataset to pandas framework

data = pd.DataFrame(housing.data)
```

```python
data.head()
```

```python
#Putting the names of the columns for better understanding of the data

data.columns=['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']
```

```python
data.head()
```

```python
target = pd.DataFrame(housing.target)
```

```python
target.head()
```

```python
target.columns=['HousePrice']
```

```python
target.head()
```

```python
data.dtypes
```

```python
print(data.isnull().sum())
```

```python
target.dtypes
```

```python
df_standardized = (data - data.mean()) / data.std()
```

```python
df_standardized.head()
```

```python
target_standardized = (target - target.mean()) / target.std()
```

```python
# Splitting features (X) and target (y)

X = df_standardized
y = target_standardized
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
df_final = pd.concat([X, y], axis=1)
df_final.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Histograms before standardization
data.hist(bins=20, figsize=(12, 8))
plt.suptitle('Histograms of Features Before Standardization', fontsize=16)
plt.show()

# 2. Boxplots to detect outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=data)
plt.title('Boxplots of Features Before Standardization')
plt.xticks(rotation=45)
plt.show()

# 3. Histograms after standardization
df_standardized.hist(bins=20, figsize=(12, 8))
plt.suptitle('Histograms of Features After Standardization', fontsize=16)
plt.show()

# 4. Correlation
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Features')
plt.show()

# 5. Scatter plots for top correlated features
correlations = data.corrwith(target['HousePrice']).sort_values(ascending=False)
top_features = correlations.index[:2]

# Scatterplots
plt.figure(figsize=(12, 6))
for i, feature in enumerate(top_features):
    plt.subplot(1, 2, i+1)
    sns.scatterplot(x=data[feature], y=target['HousePrice'])
    plt.title(f'Scatterplot: {feature} vs HousePrice')
    plt.xlabel(feature)
    plt.ylabel('HousePrice')
plt.tight_layout()
plt.show()
```

```python
train_pd, test_pd, val_pd = df_final[:17000], df_final[17000:18820], df_final[18820:]
len(train_pd), len(test_pd), len(val_pd)
```

```python
X_train, y_train = train_pd.to_numpy()[:, :-1], train_pd.to_numpy()[:, -1]
X_val, y_val = val_pd.to_numpy()[:, :-1], val_pd.to_numpy()[:, -1]
X_test, y_test = test_pd.to_numpy()[:, :-1], test_pd.to_numpy()[:, -1]

X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Linear regression
lr = LinearRegression().fit(X_train, y_train)
```

```python
# Finding the mean_absolute_error and r2 score
mae = mean_absolute_error(lr.predict(X_train), y_train)
r2 = r2_score(lr.predict(X_train), y_train)
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")
```

```python
#Finding the mean_sqared_error for both training and validation set
mse_train = mean_squared_error(lr.predict(X_train), y_train)
mse_val = mean_squared_error(lr.predict(X_val), y_val)
print(f"MSE of the training set: {mse_train}")
print(f"MSE of the validation set: {mse_val}")
```

```python
# Predictions for Linear Regression
y_train_pred_lr = lr.predict(X_train)
y_val_pred_lr = lr.predict(X_val)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred_lr, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', color='red', lw=2)
plt.title('Linear Regression - Training Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.subplot(1, 2, 2)
plt.scatter(y_val, y_val_pred_lr, alpha=0.5, color='green')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', color='red', lw=2)
plt.title('Linear Regression - Validation Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.tight_layout()
plt.show()
```

```python
#GRB
gbr = GradientBoostingRegressor(n_estimators=250).fit(X_train, y_train)
```

```python
# Finding the mean_absolute_error and r2 score
mae = mean_absolute_error(gbr.predict(X_train), y_train)
r2 = r2_score(gbr.predict(X_train), y_train)
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")
```

```python
#Finding the mean_sqared_error for both training and validation set
mse_train = mean_squared_error(gbr.predict(X_train), y_train)
mse_val = mean_squared_error(gbr.predict(X_val), y_val)
print(f"MSE of the training set: {mse_train}")
print(f"MSE of the validation set: {mse_val}")
```

```python
# Predictions for Gradient Boosting
y_train_pred_gbr = gbr.predict(X_train)
y_val_pred_gbr = gbr.predict(X_val)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred_gbr, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', color='red', lw=2)
plt.title('Gradient Boosting - Training Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.subplot(1, 2, 2)
plt.scatter(y_val, y_val_pred_gbr, alpha=0.5, color='green')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', color='red', lw=2)
plt.title('Gradient Boosting - Validation Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.tight_layout()
plt.show()
```

```python
#Adaboost
from sklearn.ensemble import AdaBoostRegressor

abr = AdaBoostRegressor(n_estimators=250).fit(X_train, y_train)
```

```python
# Finding the mean_absolute_error and r2 score
mae = mean_absolute_error(abr.predict(X_train), y_train)
r2 = r2_score(abr.predict(X_train), y_train)
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")
```

```python
#Finding the mean_sqared_error for both training and validation set
mse_train = mean_squared_error(abr.predict(X_train), y_train)
mse_val = mean_squared_error(abr.predict(X_val), y_val)
print(f"MSE of the training set: {mse_train}")
print(f"MSE of the validation set: {mse_val}")
```

```python
# Predictions for AdaBoost
y_train_pred_abr = abr.predict(X_train)
y_val_pred_abr = abr.predict(X_val)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred_abr, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', color='red', lw=2)
plt.title('AdaBoost - Training Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.subplot(1, 2, 2)
plt.scatter(y_val, y_val_pred_abr, alpha=0.5, color='green')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', color='red', lw=2)
plt.title('AdaBoost - Validation Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.tight_layout()
plt.show()
```

```python
#NN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

# Define the model
simple_nn = Sequential()
simple_nn.add(InputLayer((8,)))
simple_nn.add(Dense(2, activation='relu'))
simple_nn.add(Dense(1, activation='linear'))


opt = Adam(learning_rate=0.1)
cp = ModelCheckpoint('models/simple_nn.keras', save_best_only=True)
simple_nn.compile(optimizer=opt, loss='mse', metrics=[RootMeanSquaredError()])

simple_nn.fit(
    x=X_train,
    y=y_train,
    validation_data=(X_val, y_val),
    callbacks=[cp],
    epochs=100
)
```

```python
from tensorflow.keras.models import load_model

simple_nn = load_model('models/simple_nn.keras')
mse_train = mean_squared_error(simple_nn.predict(X_train), y_train, squared=False)
mse_val = mean_squared_error(simple_nn.predict(X_val), y_val, squared=False)
print(f"MSE of the training set: {mse_train}")
print(f"MSE of the validation set: {mse_val}")
```

```python
# Predictions for Simple Neural Network
y_train_pred_nn = simple_nn.predict(X_train).flatten()
y_val_pred_nn = simple_nn.predict(X_val).flatten()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred_nn, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', color='red', lw=2)
plt.title('Simple Neural Network - Training Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.subplot(1, 2, 2)
plt.scatter(y_val, y_val_pred_nn, alpha=0.5, color='green')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', color='red', lw=2)
plt.title('Simple Neural Network - Validation Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

plt.tight_layout()
plt.show()
```

```python
medium_nn = Sequential()
medium_nn.add(InputLayer((8,)))
medium_nn.add(Dense(32, 'relu'))
medium_nn.add(Dense(16, 'relu'))
medium_nn.add(Dense(1, 'linear'))

opt = Adam(learning_rate=.1)
cp = ModelCheckpoint('models/medium_nn.keras', save_best_only=True)
medium_nn.compile(optimizer=opt, loss='mse', metrics=[RootMeanSquaredError()])
medium_nn.fit(x=X_train, y=y_train, validation_data=(X_val, y_val), callbacks=[cp], epochs=100)
```

```python
medium_nn = load_model('models/medium_nn.keras')
mse_train = mean_squared_error(medium_nn.predict(X_train), y_train, squared=False)
mse_val = mean_squared_error(medium_nn.predict(X_val), y_val, squared=False)
print(f"MSE of the training set: {mse_train}")
print(f"MSE of the validation set: {mse_val}")
```

```python
large_nn = Sequential()
large_nn.add(InputLayer((8,)))
large_nn.add(Dense(256, 'relu'))
large_nn.add(Dense(128, 'relu'))
large_nn.add(Dense(64, 'relu'))
large_nn.add(Dense(32, 'relu'))
large_nn.add(Dense(1, 'linear'))

opt = Adam(learning_rate=.1)
cp = ModelCheckpoint('models/large_nn.keras', save_best_only=True)
large_nn.compile(optimizer=opt, loss='mse', metrics=[RootMeanSquaredError()])
large_nn.fit(x=X_train, y=y_train, validation_data=(X_val, y_val), callbacks=[cp], epochs=100)
```

```python
large_nn = load_model('models/large_nn.keras')
mse_train = mean_squared_error(large_nn.predict(X_train), y_train, squared=False),
mse_val = mean_squared_error(large_nn.predict(X_val), y_val, squared=False)
print(f"MSE of the training set: {mse_train}")
print(f"MSE of the validation set: {mse_val}")
```

```python
#best model - GBR
mse_test = mean_squared_error(gbr.predict(X_test), y_test)
print(f"MSE of the testing set: {mse_test}")
```

Thus we can conclude that the Gradient Boost regressor was by far the more accurate and more reliable one as it recorded the lowest MAE (0.264) and the highest R score of 0.812 and can predict the target variable to the extent of 81.2% whereby reasonable expectations can be placed on the predictive power of this model.
